```python
def get_datetime(Date):
    date_string = str(Date)
    my_dict = {}
    my_dict['year'] = date_string[:4] 
    my_dict['month'] = date_string[4:6]
    my_dict['day'] = date_string[6:8]
    return my_dict['day']
```

```python

my_date = 20120530
mydate_dict = get_datetime(my_date)
```
